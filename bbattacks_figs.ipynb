```python
# MI-FGSM, iCARL, VGG19
x_values = np.arange(10, 110, 10)
y_values = [32.33, 31.05, 28.26, 25.65, 24.12, 25.0, 23.23, 21.22, 21.41, 22.32]

plt.figure(figsize=(3.5, 3.5))  # Adjusted for IEEE one-column format
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')

plt.xlabel('Adversarial Samples (%)', fontsize=14)
# plt.ylabel('Accuracy (%)', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.title('Accuracy Drop', fontsize=10)

plt.grid(True)
plt.tight_layout()
# plt.savefig('mifgsm_icarl_vgg19.pdf', format='pdf', dpi=400)
plt.savefig('mifgsm_icarl_vgg19.png', dpi=400)

# Display the plot
plt.show()
```

```python
# DST-MI-FGSM, Replay, ResNet18
x_values = np.arange(10, 110, 10)
y_values = [32.59, 28.48, 28.34, 26.44, 23.3, 22.74, 23.95, 23.55, 22.16, 21.91]

plt.figure(figsize=(3.5, 3.5))  # Adjusted for IEEE one-column format
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')

plt.xlabel('Adversarial Samples (%)', fontsize=14)
plt.ylabel('Accuracy (%)', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.title('Accuracy Drop', fontsize=10)

plt.grid(True)
plt.tight_layout()
# plt.savefig('dstmifgsm_replay_resnet18.pdf', format='pdf', dpi=400)
plt.savefig('dstmifgsm_replay_resnet18.png', dpi=400)

# Display the plot
plt.show()
```
